```python
lines = """#isittime  
www.whatismyname.com  
#letusgo""".split()
```

```python
import re
pat = re.compile(r'^(#([a-z0-9.]+))|([a-z0-9]+\.([a-z0-9]+)\.[a-z0-9]+)$')
texts = []
for line in lines:
    _, htext, _, dtext = pat.findall(line)[0]
    text = htext or dtext
    texts.append(text.lower())
```

```python
from collections import defaultdict
word_hash = defaultdict(list)
for w in open("data/words.txt"):
    w = w.strip().lower()
    h = reduce(lambda h, c: h*127+ord(c), w, 0) % 5000
    word_hash[h].append(w)
```

```python
word_hash = set([w.strip().lower() for w in open("data/words.txt").readlines()])
```

```python
def parse_words(tag, word_hash):
    words = []
    s = 0
    while s < len(word_hash):
        for i in range(s+1, len(tag)+1)[::-1]:
            if tag[s:i] in word_hash:
                words.append(tag[:i])
                s = i+1
    return words
```

```python
parse_words("whatmyname", word_hash)
```
