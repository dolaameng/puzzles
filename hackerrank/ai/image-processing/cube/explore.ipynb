```python
import matplotlib.pyplot as plt
import numpy as np 
import scipy as sp
from scipy import ndimage, misc, spatial
from sklearn.feature_extraction import image
from sklearn import cluster
```

```python
img_solved = misc.imread("data/cube1solved.jpg")
img_unsolved = misc.imread("data/cube2unsolved.jpg")
#img = img_unsolved
img = img_solved
```

```python
## scaling should be enough
scaled = img[::4, ::4]
## convert to gray -- for edge detection
gray = 0.299 * scaled[:,:,0] + 0.587 * scaled[:,:,1] + 0.114 * scaled[:,:,2]
```

```python
## finding main edges
#edges = ndimage.filters.uniform_filter(gray, (4, 4))
edges = ndimage.filters.sobel(gray)
edges = np.where(edges > 120, 1, 0)
edge_points = np.asarray(zip(*np.where(edges > 0)))
```

```python
## find convex hull of the edge points
hull = spatial.ConvexHull(edge_points)
## and thus find the contours
contour = np.zeros_like(edges)
for simplex in hull.simplices:
    pts = edge_points[simplex]
    r0, r1 = pts[:, 0]
    c0, c1 = pts[:, 1]
    l = int(np.hypot(r1-r0, c1-c0))
    rs, cs = np.linspace(r0, r1, l), np.linspace(c0, c1, l)
    contour[rs.astype(np.int), cs.astype(np.int)] = 1
```

```python
plt.imshow(edges, cmap = plt.cm.gray)
plt.autoscale(False)
for simplex in hull.simplices:
    plt.plot(edge_points[simplex, 1], edge_points[simplex, 0], "ro-")
```

```python
plt.imshow(contour, cmap = plt.cm.gray)
```

```python
## fill the inside of contour
filled = np.zeros_like(edges)
nr, nc = filled.shape
for r in xrange(nr):
    mask = 0
    for c in xrange(nc):
         # stride at the boundary - rotate the color
        if contour[r, c] == 1 and contour[r, c-1] == 0:
            mask = (mask + 1) % 2
        filled[r, c] += mask
```

```python
plt.imshow(filled, cmap = plt.cm.gray)
```

```python
## get the shadow - back to RGB, so that it can be combined with original
shadowed = filled[:, :, np.newaxis]
shadowed = np.concatenate([shadowed, shadowed, shadowed], axis = 2)
shadowed = scaled + shadowed * 45 # make the filled area brighter- remove the dark line
#shadowed = scaled * shadowed
## make it blur and smooth by max and uniform fitler
max_img = ndimage.filters.maximum_filter(shadowed, (5, 5, 1))
max_img = ndimage.filters.uniform_filter(max_img, (4, 4, 1))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(scaled)
axes[1].imshow(max_img)
```

```python
## do clustering on colors
nr,nc,_ = max_img.shape
connectivity = image.grid_to_graph(n_x = nr, n_y=nc)
n_clusters = 4
ward = cluster.AgglomerativeClustering(n_clusters=n_clusters, 
                                       connectivity=connectivity, 
                                       linkage="ward").fit(max_img.reshape(-1, 3))
label = ward.labels_.reshape(nr, nc)
```

```python
plt.imshow(label)
```

```python
## analyze the centers of clusters
centers = [ndimage.measurements.center_of_mass(label==l) 
           for l in np.unique(label) if l !=label[0,0]]
```

```python
dist01 = spatial.distance.euclidean(centers[0],centers[1])
dist12 = spatial.distance.euclidean(centers[1],centers[2])
dist23 = spatial.distance.euclidean(centers[2],centers[0])
```

```python
dist01, dist12, dist23
```

```python
mindist = 0.8 * (dist01+dist23+dist12)/3
maxdist = 1.2 * (dist01+dist23+dist12)/3

if (dist01>mindist and dist12>mindist and dist23>mindist \
        and dist01<maxdist and dist12<maxdist and dist23<maxdist):
    print 'solved'
else:
    print 'unsolved'
```
