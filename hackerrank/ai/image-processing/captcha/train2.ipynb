```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
from collections import defaultdict
import pickle as pkl
import base64
import zlib
```

```python
out_files = ["output/output%02i.txt" % i for i in xrange(0, 25)]
in_files = ["input/input%02i.jpg" % i for i in xrange(0, 25)]
```

```python
print zip(in_files, out_files)
```

```python
## segment letters
img_dict = defaultdict(list)
for fin, fout in zip(in_files, out_files):
    try:
        img = plt.imread(fin)
        img = (img <= 15)[:,:,0]
        img_segs = [img[11:21, 5+i*9:13+i*9] for i in xrange(5)]
        letters = open(fout).readline().strip()
        for l, s in zip(letters, img_segs):
            img_dict[l].append(s)
    except Exception, e:
        ##print e
        pass
```

```python
img_dict = dict([(k, np.dstack(img_dict[k]).mean(axis = 2)) for k in img_dict.keys()])
```

```python
## test case
t = img_dict["O"]
dist_dict = dict([ (letter, np.mean((t-img)**2)) for letter, img in img_dict.iteritems()])
```

```python
sorted(map(lambda (k,v): (k,np.mean(v)), dist_dict.items()), key = lambda (l, d): d)
```

```python
s = base64.encodestring(zlib.compress(pkl.dumps(img_dict), 9))
```

```python
s
```

```python
model = pkl.loads(zlib.decompress(base64.decodestring(s)))
```

```python
def predict_seg(seg, model):
    dist_dict = dict([ (letter, np.mean((seg-img)**2)) for letter, img in model.iteritems()])
    avg_dist = sorted(dist_dict.items(), key = lambda (l, d): d)
    return avg_dist[0][0]

def predict_img(img, model):
    img = (img <= 15)[:,:,0]
    img_segs = [img[11:21, 5+i*9:13+i*9] for i in xrange(5)]
    return "".join([predict_seg(s, model) for s in img_segs])
```

```python
## test result
test_files = ["input/input%02i.txt" % i for i in xrange(0, 25)]
images = [np.asarray([ [map(int, pixel.split(",")) 
                            for pixel in line.strip().split(" ")] 
                            for line in open(tfile).readlines()[1:]]) for tfile in test_files]
predictions = [predict_img(img, model) for img in images]
```
