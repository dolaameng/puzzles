```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
from scipy import ndimage as nd
```

```python
img_file = "input/input02.txt"
lines = open(img_file).readlines()
nrow, ncol = map(int, lines[0].split(" "))
print nrow, ncol
img1 = np.asarray([ [map(int, pixel.split(",")) 
                            for pixel in line.strip().split(" ")] 
                            for line in lines[1:]])
print img1.shape
plt.imshow(img1, interpolation="nearest" )
```
