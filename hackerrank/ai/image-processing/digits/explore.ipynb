```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage, misc
```

```python
def parse_img(bytes):
    lines = map(lambda s: s.strip(), bytes.strip().split("\n"))
    nr, nc = map(int, lines[0].split(" "))
    #img = np.zeros((nr, nc, 3))
    img = np.array([[rgb.split(",") for rgb in line.split(" ")] 
                    for line in lines[1:]], dtype = np.int)
    return img
```

```python
s = open("data/input11.txt").read()
img = parse_img(s)
print img.shape
```

```python
plt.imshow(img)
plt.figure()
plt.imshow(misc.imresize(img, size = (120, 60, 3)))
plt.figure()
plt.imshow(ndimage.filters.gaussian_filter(img, 3))
```
