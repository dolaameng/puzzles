```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage, misc
```

```python
## extract digit images from training data
digits = misc.imread("data/digits.jpg")[:,:,0]
digit_imgs = {}
for i, x in enumerate(np.arange(0, 528, 55)):
    digit_imgs[i] = digits[:, x:(34+x)]
    digit_imgs[i] = digit_imgs[i] < digit_imgs[i].mean()
    digit_imgs[i] = misc.imresize(digit_imgs[i], (57, 34))
    plt.figure()
    plt.imshow(digit_imgs[i], cmap = plt.cm.gray)
```
