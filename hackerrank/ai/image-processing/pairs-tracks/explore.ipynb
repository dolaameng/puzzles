```python
def binary_and_open(sig):
    row,col = sig.shape[:2]
    sigmax = sig.max()
    sigmin = sig.min()
    for i in range(row):
        for j in range(col):
            sig[i,j] = (sig[i,j] - sigmin)/(sigmax-sigmin)*255.0
            if sig[i,j] < 170: sig[i,j] = 0
            else: sig[i,j] = 1
    return ndimage.binary_opening(sig, structure=np.ones((10,10))).astype(np.int)
```

```python
bytes = StringIO(urlopen("http://ww2.valdosta.edu/~algriepentrog/circlergb.jpg").read())
img = misc.imread(bytes)
img = misc.imresize(img, (120,120))
```

```python
gray = img[:,:,0] * .2126 + img[:,:,1] * .7152 + img[:,:,2] * .0722
plt.imshow(gray, cmap = plt.cm.gray)
```

```python
binary = (gray > 120)
binary = ndimage.filters.uniform_filter(binary, 3)
binary = ndimage.filters.sobel(binary)
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
rcs = zip(*np.where(binary > 0))
hull = spatial.ConvexHull(rcs)
```

```python
for simplex in hull.simplices:
    p0,p1=rcs[simplex[0]], rcs[simplex[1]]
    plt.plot([p0[0],p1[0]], [p0[1],p1[1]], "r-")
```
