```python
## solution 1
@solution_url
def solution1():
    word = "map"
    letters = string.letters
    letters_shifted = letters[2:] + letters[:2]
    tab = string.maketrans(letters, letters_shifted)
    return string.translate(word, tab)

print solution1()
```

```python
## solution 2
@solution_url
def solution2(txt_file = 'data/problem3.txt'):
    text = open(txt_file).read()
    od = OrderedDict()
    for c in text:
        od[c] = od.get(c, 0) + 1
    return ''.join([c for c in od if od[c] == 1])

print solution2()
```

```python
## solution3
@solution_url
def solution3(f = "data/problem4.txt"):
    text = open(f).read()
    pat = re.compile(r"[^A-Z][A-Z]{3}([a-z])[A-Z]{3}[^A-Z]")
    return ''.join(pat.findall(text))

print solution3()
```
