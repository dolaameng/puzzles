```python
## solution 2
@solution_url
def solution2(txt_file = 'data/problem3.txt'):
    text = open(txt_file).read()
    od = OrderedDict()
    for c in text:
        od[c] = od.get(c, 0) + 1
    return ''.join([c for c in od if od[c] == 1])

print solution2()
```

```python
## solution 14
img = pl.imread('data/wire.png')
print img.shape
pl.imshow(img.reshape(100, 100, 3)) ## GOT you took the wrong curve
## the word curve and the Italy roll made a strong suggestion of a spiral layout
## the equation (hint) 100*100 = (100+99+99+98) +(... suggested that the spiral
## started from outside

## IT TOOK ME LONG TIME TO MAKE THE GENERATOR CORRECT and CONCISE
## ideas borrowed from http://unixwars.com/2007/09/20/python-challenge-level-14-walk-around/
def spiral(steps = 100):
    rc = (0, -1) # left-top clockwise spiral
    headings = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    double_step = steps * 2
    while True:
        for h in headings:
            for _ in xrange(double_step // 2):     
                rc = map(sum, zip(rc, h))
                yield rc
            double_step -= 1
        
        
spiral_series = spiral()
new_img = np.zeros((100, 100, 3))
for i in xrange(100*100):
    r, c = spiral_series.next()
    new_img[r, c, :] = img[0, i, :]
    
pl.figure()
pl.imshow(new_img)

print 'http://www.pythonchallenge.com/pc/return/%s.html' % 'cat'
print 'http://www.pythonchallenge.com/pc/return/%s.html' % 'uzi'
```
